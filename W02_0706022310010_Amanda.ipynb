1. For bedrooms, bathrooms, and floors, we used the mode (most frequent value) because these columns are about counts of rooms or floors. The mode makes sense since it represents the most common situation for properties.
2. For land_size, building_age, and building_size, we used the median because these columns have large numeric ranges. The median is less affected by very small or very large values compared to the mean, so it gives a fairer replacement for missing values.
3. For the categorical columns certificate, property_condition, and furnishing, the missing values were replaced with â€œunknownâ€. This way, no data is lost, and we can still capture the information that the value was not provided, instead of dropping the rows or filling with something misleading.

```python
# Fill nulls for continuous numeric columns with mode
for col in ['bedrooms', 'bathrooms', 'floors']:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Fill nulls for continuous numeric columns with median
for col in ['land_size_m2', 'building_size_m2','year_built']:
    df[col].fillna(df[col].median(), inplace=True)

# Fill nulls for categorical columns with unknown
for col in ['certificate', 'property_condition', 'furnishing']:
    df[col] = df[col].fillna('unknown')
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)

# Show result or message if none
if null_df.empty:
    print("No null data")
else:
    print(null_df)
```

## BINNING

1. Buyers and sellers almost always filter properties by budget. Binning price into ranges makes it easier to compare affordability segments and spot which price ranges dominate the dataset.
2. Categorizing by era (mid-century, late 20th century, modern) makes more sense than treating each year individually, since buyers and analysts usually care about â€œnew vs oldâ€ rather than exact build years.
3. Buyers often search based on lot size, and segmenting into small, standard, medium, large, and estate categories reflects how properties are marketed in real estate.
4. Buyers often search based on lot size, and segmenting into small, standard, medium, large, and estate categories reflects how properties are marketed in real estate.

```python
# PRICE
bins = [0, 1e9, 4e9, 20e9, np.inf] # Bins: 0-1B, 1B-4B, 4B-20B, >20B
labels = ['Affordable (<1B)', 'Mid-Range (1B-4B)', 'Expensive (4B-20B)', 'Very Expensive (>20B)']

df['price_category'] = pd.cut(df['price_in_rp'], bins=bins, labels=labels, right=False)

print("\nCounts per price_category:")
print(df['price_category'].value_counts())

# YEAR
bins = [ 1945, 1980, 2000, 2022]
labels = ['Mid-Century (1946-1980)', 'Late 20th Century (1981-2000)', 'Modern (2001-2022)']

df['building_era'] = pd.cut(df['year_built'], bins=bins, labels=labels)

print("\nCounts per building_era:")
print(df['building_era'].value_counts())
```

```python
# LAND SIZE
land_bins = [0, 100, 250, 500, 1000, np.inf]
land_labels = ['Small Lot (<100 mÂ²)', 'Standard Lot (101-250 mÂ²)', 'Medium Lot (251-500 mÂ²)', 'Large Lot (501-1000 mÂ²)', 'Estate (>1000 mÂ²)']

df['land_category'] = pd.cut(df['land_size_m2'], bins=land_bins, labels=land_labels)

print("\nCounts per land_category:")
print(df['land_category'].value_counts())

# BUILDING SIZE
building_bins = [0, 75, 150, 300, 600, np.inf]
building_labels = ['Compact (<75 mÂ²)', 'Standard (76-150 mÂ²)', 'Large (151-300 mÂ²)', 'Very Large (301-600 mÂ²)', 'Mansion (>600 mÂ²)']

df['building_category'] = pd.cut(df['building_size_m2'], bins=building_bins, labels=building_labels)

print("\nCounts per building_size:")
print(df['building_category'].value_counts())
```

## Exploratory Data Analysis (EDA)

1. I want to analyze how the location of a property affects house prices. This will help reveal whether certain areas have higher prices.
2. I want to analyze how a property's furnishing status affects house prices. This will help reveal whether certain status have higher prices.
3. I want to focus on the top 10 most common facilities in the dataset to create a cleaner table and a more meaningful bar chart that shows the facilties most houses have.
4. I want to analyze how property conditions vary across different cities. This will help reveal whether certain cities tend to have better-maintained or newer houses compared to others.
5. I want to analyze whether the year a property was built has any relationship with the number of bedrooms, bathrooms, floors, garages, and carports. This will help reveal if newer or older houses tend to have more or fewer of these features.

```python
# house price x city
plt.figure(figsize=(12, 7))
sns.boxplot(x='city', y='price_in_rp', data=df)

plt.title('House Price Distribution by City', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Price in Rupiah (in Billions)', fontsize=12)

formatter = lambda x, pos: f'{x/1e9:.1f}B'
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()

# house price x furnishing
plt.figure(figsize=(12, 7))
sns.boxplot(x='furnishing', y='price_in_rp', data=df)

plt.title('House Price Distribution by furnishing', fontsize=16)
plt.xlabel('furnishing', fontsize=12)
plt.ylabel('Price in Rupiah (in Billions)', fontsize=12)

formatter = lambda x, pos: f'{x/1e9:.1f}B'
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()

# the count of each facilities
all_facilities = df['facilities'].str.split(',').explode().str.strip()

top_10_facilities = all_facilities.value_counts().head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_facilities.values, y=top_10_facilities.index, palette='viridis')

plt.title('Most Common House Facilities', fontsize=16)
plt.xlabel('Number of Houses', fontsize=12)
plt.ylabel('Facility', fontsize=12)

plt.show()

# to see city x property_condition
cross_tab = pd.crosstab(df['city'], df['property_condition'])
proportions = cross_tab.div(cross_tab.sum(axis=1), axis=0)
ax = proportions.plot(kind='bar',
                      stacked=True,
                      figsize=(10, 7),
                      colormap='viridis')

plt.title('Distribution of Property Condition by City', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Proportion', fontsize=12)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
plt.xticks(rotation=0)
plt.legend(title='property_condition', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# year built correlation to the house (bedrooms etc)
selected_features = ['year_built', 'bedrooms', 'bathrooms', 'floors', 'garages', 'carports']
df_selected = df[selected_features]

correlation_matrix = df_selected.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,
            annot=True,      
            cmap='coolwarm', 
            fmt='.2f',    
            linewidths=.5)

plt.title('Correlation Heatmap of Housing Features', fontsize=16)
plt.show()
```
